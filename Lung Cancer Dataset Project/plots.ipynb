```python
import pandas as pd

data = pd.read_csv('dataset.csv')

def dataset_overview(df):
    data_info = df.dtypes
    missing_values = df.isnull().sum()
    unique_values = df.nunique()
    overview = pd.DataFrame({
        'Data Type': data_info,
        'Missing Values': missing_values,
        'Unique Values': unique_values
    })
    return overview

overview = dataset_overview(data)
print("Dataset Overview:")
print(overview)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data setup
data = {
    'Patient Id': [f'P{i}' for i in range(1, 1001)],
    'Age': np.random.randint(20, 80, size=1000),
    'Gender': np.random.choice([0, 1], size=1000),
    'Air Pollution': np.random.randint(1, 9, size=1000),
    'Alcohol use': np.random.randint(1, 9, size=1000),
    'Dust Allergy': np.random.randint(1, 9, size=1000),
    'OccuPational Hazards': np.random.randint(1, 9, size=1000),
    'Genetic Risk': np.random.randint(1, 8, size=1000),
    'chronic Lung Disease': np.random.randint(1, 8, size=1000),
    'Balanced Diet': np.random.randint(1, 8, size=1000),
    'Obesity': np.random.randint(1, 8, size=1000),
    'Smoking': np.random.randint(1, 9, size=1000),
    'Passive Smoker': np.random.randint(1, 9, size=1000),
    'Chest Pain': np.random.randint(1, 10, size=1000),
    'Coughing of Blood': np.random.randint(1, 10, size=1000),
    'Fatigue': np.random.randint(1, 9, size=1000),
    'Weight Loss': np.random.randint(1, 9, size=1000),
    'Shortness of Breath': np.random.randint(1, 9, size=1000),
    'Wheezing': np.random.randint(1, 9, size=1000),
    'Swallowing Difficulty': np.random.randint(1, 9, size=1000),
    'Clubbing of Finger Nails': np.random.randint(1, 10, size=1000),
    'Frequent Cold': np.random.randint(1, 8, size=1000),
    'Dry Cough': np.random.randint(1, 8, size=1000),
    'Snoring': np.random.randint(1, 8, size=1000),
    'Level': np.random.choice(['Low', 'Medium', 'High'], size=1000)
}

df = pd.DataFrame(data)

# 1. Bar Plot of Unique Values in Categorical Features
categorical_columns = df.select_dtypes(include='int64').columns
n_cols = 4  # You can adjust this to a suitable number
n_rows = (len(categorical_columns) // n_cols) + 1  # Calculate number of rows needed

plt.figure(figsize=(15, 6))
for i, col in enumerate(categorical_columns, 1):
    plt.subplot(n_rows, n_cols, i)
    df[col].value_counts().plot(kind='bar', title=col, color='skyblue')
    plt.tight_layout()

plt.show()

# 2. Pie Chart of "Level" Distribution
plt.figure(figsize=(7, 7))
df['Level'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Distribution of Level')
plt.ylabel('')
plt.show()

# 3. Histogram of Age
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='green')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 4. Stacked Bar Plot of Common Symptoms vs "Level"
symptoms = ['Chest Pain', 'Fatigue', 'Shortness of Breath', 'Wheezing', 'Coughing of Blood']
df_symptoms = df[symptoms + ['Level']]

df_symptoms_grouped = df_symptoms.groupby('Level').sum()
df_symptoms_grouped.plot(kind='bar', stacked=True, figsize=(10, 6), cmap='tab10')
plt.title('Stacked Bar Plot: Symptoms vs Level')
plt.ylabel('Count of Symptoms')
plt.show()

# 5. Heatmap of Missing Values
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```
